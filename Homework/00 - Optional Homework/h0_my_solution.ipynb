```python
import os
path = os.path.join(os.getcwd(), "data")
df = pd.read_csv(f"{path}/interactive_data.csv")
df
```

```python
# filter
df_filtered = df.loc[~(df=='None selected').any(axis=1)]
df_filtered
```

```python
df_filtered['Age'].unique()
```

```python
# results
all_deaths = df_filtered.sum(axis=0)['Deaths']

df_suicides = df_filtered.loc[(df_filtered=='Suicide').any(axis='columns')]
suicides = df_suicides.sum(axis='index')['Deaths']
print(f"Percentage of suicides: {100*suicides / all_deaths}%")

males = df_filtered.loc[(df_filtered=='Male').any(axis='columns')].sum(axis='index')['Deaths']
male_suicides = df_suicides.loc[(df_suicides=='Male').Gender].sum(axis='index')['Deaths']
print(f"Percentage of male suicides: {100*male_suicides / suicides}%")

df_homicides = df_filtered.loc[(df_filtered=='Homicide').Intent]
homicides = df_homicides.sum(axis=0).Deaths
print(f"Percentage of homicides: {(100 * homicides / all_deaths)}%")

df_homicides_male_between_15_and_34 = df_homicides.loc[(df_homicides['Gender']=='Male')
                                                       & (df_homicides['Age']=='15 - 34')]
homicides_male_between_15_and_34 = df_homicides_male_between_15_and_34.sum(axis=0).Deaths
df_black_male_homicides_between_15_and_34 = df_homicides.loc[(df_homicides['Gender']=='Male')
                                                            & (df_homicides['Race']=='Black')
                                                            & (df_homicides['Age']=='15 - 34')]
black_male_homicide_between_15_and_34 = df_black_male_homicides_between_15_and_34.Deaths.sum(axis=0)

print(f"Percentage of males homicides with black colour between 15 and 34: {100 * black_male_homicide_between_15_and_34 / homicides_male_between_15_and_34}%")

female_homicides = df_filtered.loc[(df_filtered.Gender=='Female')
                               & (df_filtered.Intent=='Homicide')].Deaths.sum(axis=0)
print(f"Percentage of female homicides {100 * female_homicides / homicides}%")

females = df_filtered.loc[(df_filtered=='Female').any(axis='columns')].sum(axis='index')['Deaths']
```
